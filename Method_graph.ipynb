```python
import graphviz as gv
%load_ext jupyter_black
```

```python
dot = gv.Digraph(
    "Figure_method",
    comment="Figure_method",
    format="png",
)
dot.attr(rankdir="LR")

# Create nodes
dot.node("E-fuel impact", shape="rectangle")
dot.node("Other source \nof impact", shape="rectangle")
dot.node("H2 impact", shape="rectangle")
dot.node("Electricity impact", shape="rectangle")
dot.node("Electricity mix\nconsidered", shape="rectangle")
dot.node("Production mix", shape="rectangle")
dot.node("Tracked consumption mix", shape="rectangle")
dot.node("Grid losses", shape="rectangle")
dot.node("Tracking algorithm", shape="rectangle")
dot.node("ENTSO-E\nproduction data", shape="rectangle")
dot.node("ENTSO-E\nexchange data", shape="rectangle")

dot.node("Emissions factors \nper electricity output", shape="rectangle")
dot.node("Emissions factors \nper fuel input", shape="rectangle")
dot.node("Efficiency data", shape="rectangle")
dot.node("EUROSTAT\nelectricity and heat output", shape="rectangle")
dot.node("EUROSTAT\nenergy input", shape="rectangle")
dot.node("EUROSTAT\ndataset nrg_bal_c", shape="rectangle")
dot.node("EUROSTAT\ndataset nrg_cb_e", shape="rectangle")

# Create links
dot.edge("H2 impact", "E-fuel impact", style="dashed")
dot.edge("Other source \nof impact", "E-fuel impact", style="dashed")
dot.edge("Electricity impact", "H2 impact")
dot.edge("Electricity mix\nconsidered", "Electricity impact")
dot.edge("Emissions factors \nper electricity output", "Electricity impact")

dot.edge(
    "Emissions factors \nper fuel input", "Emissions factors \nper electricity output"
)
dot.edge("Efficiency data", "Emissions factors \nper electricity output")

dot.edge("EUROSTAT\nelectricity and heat output", "Efficiency data")
dot.edge("EUROSTAT\nenergy input", "Efficiency data")

dot.edge("Production mix", "Electricity mix\nconsidered", style="dashed")  # or
dot.edge("Tracked consumption mix", "Electricity mix\nconsidered", style="dashed")
dot.edge("Grid losses", "Electricity mix\nconsidered")

dot.edge("EUROSTAT\ndataset nrg_cb_e", "Grid losses")

dot.edge("ENTSO-E\nproduction data", "Production mix")

dot.edge("Tracking algorithm", "Tracked consumption mix")

dot.edge("ENTSO-E\nproduction data", "Tracking algorithm")
dot.edge("ENTSO-E\nexchange data", "Tracking algorithm")

dot.edge("EUROSTAT\ndataset nrg_bal_c", "EUROSTAT\nelectricity and heat output")
dot.edge("EUROSTAT\ndataset nrg_bal_c", "EUROSTAT\nenergy input")

dot.render(directory="plot")
dot.save()
```
